```python
import pandas as pd
import re
import sys
from pathlib import Path

# 1. Setup MASRI Path & Tokenizer
masri_root = str(Path(".").resolve() / "MASRI")
if masri_root not in sys.path:
    sys.path.append(masri_root)

try:
    from masri.tokenise.tokenise import MTWordTokenizer
    tokenizer = MTWordTokenizer()
    print("MASRI Tokenizer loaded successfully!")
except ImportError as e:
    print(f"Failed to load MASRI: {e}")

# 2. Combined Cleaning Logic
# Merged: punctuation, quotes, special symbols, digits, underscores, and backticks
combined_remove_regex = r'[\,\?\.\!\;\:\"\“\%\‘\”\\»\«\d\_\`\–]'
mapping = {'á': 'a', 'é': 'e', 'ć': 'ċ', 'í': 'i', 'ó': 'o', 'ʼ': "'", "’": "'"}

def full_process_maltese(text):
    if not isinstance(text, str):
        return ""
    
    # Step A: Remove special chars/digits and lowercase
    # This covers both chars_to_remove_regex sets from your snippets
    clean_text = re.sub(combined_remove_regex, '', text).lower()

    # Step B: Apply Normalization Mapping
    for char, replacement in mapping.items():
        clean_text = clean_text.replace(char, replacement)

    # Step C: MASRI Tokenization (Handles clitics and Maltese specifics)
    # tokens = tokenizer.tokenize(clean_text)
    final_text = " ".join(clean_text.split())
    
    # Step D: Re-join into a clean string
    return final_text
```

```python
# 3. Load and Process CSV
DATA_ROOT = Path.home() / "MalteseProject" 

# Load the dataset
file_path = Path("/home/ian/MalteseProject/train_all.tsv")

if not file_path.exists():
    print(f"Error: File not found at {file_path}")
else:
    # Use sep='\t' for TSV files
    df = pd.read_csv(file_path, sep='\t')

    # Apply processing to the 'text' column
    if 'text' in df.columns:
        df['processed_text'] = df['text'].apply(full_process_maltese)

        # 4. Check Vocab and Characters
        all_text = " ".join(df['processed_text'])
        vocab = sorted(list(set(all_text)))

        print(f"\n--- Results ---")
        print(f"Total rows processed: {len(df)}")
        print(f"Vocabulary Size: {len(vocab)}")
        print(f"Characters found: {vocab}")

        # 5. Create the dictionary for verification
        vocab_dict = {v: k for k, v in enumerate(vocab)}
    else:
        print(f"Error: Column 'text' not found. Columns available: {df.columns.tolist()}")
```

## Corpus Vocab

```python
import sys
from pathlib import Path

# Path to your 5GB corpus
CORPUS_PATH = Path("/home/ian/MalteseProject/corpus_untokenised.txt")

def get_corpus_vocab(file_path):
    unique_chars = set()
    line_count = 0
    
    print(f"Opening {file_path}...")
    
    try:
        # Opening with buffering to handle large file stream
        with open(file_path, 'r', encoding='utf-8') as f:
            for line in f:
                # Update the set with unique characters from the current line
                unique_chars.update(line)
                
                # Progress tracker (optional)
                line_count += 1
                if line_count % 1000000 == 0:
                    print(f"Processed {line_count} lines...")

        # Sort the final vocabulary
        vocab = sorted(list(unique_chars))
        
        print(f"\n--- Results ---")
        print(f"Total lines processed: {line_count}")
        print(f"Vocabulary Size (Characters): {len(vocab)}")
        print(f"Characters found: {vocab}")
        
        # Create the mapping for verification
        vocab_dict = {char: idx for idx, char in enumerate(vocab)}
        return vocab_dict

    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    vocab_mapping = get_corpus_vocab(CORPUS_PATH)
```
