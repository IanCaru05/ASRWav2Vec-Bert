```python
import pandas as pd
import re
import sys
from pathlib import Path

# 1. Setup MASRI Path & Tokenizer
masri_root = str(Path(".").resolve() / "MASRI")
if masri_root not in sys.path:
    sys.path.append(masri_root)

try:
    from masri.tokenise.tokenise import MTWordTokenizer
    tokenizer = MTWordTokenizer()
    print("MASRI Tokenizer loaded successfully!")
except ImportError as e:
    print(f"Failed to load MASRI: {e}")

# 2. Combined Cleaning Logic
# Merged: punctuation, quotes, special symbols, digits, underscores, and backticks
combined_remove_regex = r'[\,\?\.\!\;\:\"\“\%\‘\”\\»\«\d\_\`\–]'
mapping = {'á': 'a', 'é': 'e', 'ć': 'ċ', 'í': 'i', 'ó': 'o', 'ʼ': "'", "’": "'"}

def full_process_maltese(text):
    if not isinstance(text, str):
        return ""
    
    # Step A: Remove special chars/digits and lowercase
    # This covers both chars_to_remove_regex sets from your snippets
    clean_text = re.sub(combined_remove_regex, '', text).lower()

    # Step B: Apply Normalization Mapping
    for char, replacement in mapping.items():
        clean_text = clean_text.replace(char, replacement)

    # Step C: MASRI Tokenization (Handles clitics and Maltese specifics)
    tokens = tokenizer.tokenize(clean_text)
    
    # Step D: Re-join into a clean string
    return " ".join(tokens)
```

```python
# 3. Load and Process CSV
DATA_ROOT = Path.home() / "MalteseProject" 

# Load the dataset
file_path = DATA_ROOT / "train_all - train_all.csv"
df = pd.read_csv(file_path)

# Apply processing to the 'text' column
df['processed_text'] = df['text'].apply(full_process_maltese)

# 4. Check Vocab and Characters
all_text = " ".join(df['processed_text'])
vocab = sorted(list(set(all_text)))

print(f"\n--- Results ---")
print(f"Total rows processed: {len(df)}")
print(f"Vocabulary Size: {len(vocab)}")
print(f"Characters found: {vocab}")

# 5. Create the dictionary for verification
vocab_dict = {v: k for k, v in enumerate(vocab)}
```
