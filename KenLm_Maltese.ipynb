## Pre Processing

### Tokenised Approach

```python
import re

normalization_mapping = {
    'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ć': 'ċ', 
    '’': "'", '`': "'"  # Standardise all apostrophe types to the simple '
}

def tokensied_approach(text):
    if not isinstance(text, str):
        return ""
        
    # Initial normalization & lowercase
    text = text.lower()
    for char, replacement in normalization_mapping.items():
        text = text.replace(char, replacement)
    
    # 1. Tokenising
    tokens = tokenizer.tokenize(text)
    
    processed_tokens = []
    
    for token in tokens:
        # 2. CONVERT NUMBERS
        # Check if the token is numeric (e.g., "12" or "15.5")
        if re.match(r'\d+', token):
            try:
                word_val = num2text(int(token)) 
                processed_tokens.append(word_val)
            except:
                continue 
        
        # 3. CLEAN TOKENS
        else:
            # STRICKER REGEX: Only allows the 38 chars from W2vBert
            # Specifically: a-z, ċ, ġ, ħ, ż, à, è, ì, ò, ù, ', and -
            clean_token = re.sub(r"[^a-zċġħżàèìòù'\-]", '', token)
            
            if clean_token:
                processed_tokens.append(clean_token)

    # 4. FINAL JOIN
    # KenLM requires one line per sentence/phrase, space-separated
    return " ".join(processed_tokens)
```

### Untokenised Approach

```python
import re

normalization_mapping = {
    'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ć': 'ċ', 
    '’': "'", '`': "'"
}

def regex_approach(text):
    if not isinstance(text, str):
        return ""
        
    # 1. Lowercase and basic normalisation
    text = text.lower()
    for char, replacement in normalization_mapping.items():
        text = text.replace(char, replacement)

    # 2. Convert numbers directly in the string
    def replace_num(match):
        try:
            return num2text(int(match.group()))
        except:
            return "" 

    text = re.sub(r'\d+', replace_num, text)

    # 3. Clean characters
    text = re.sub(r"[^a-zċġħżàèìòù'\-\s]", '', text)

    # 4. Final Cleanup
    # Collapse multiple spaces into one and strip leading/trailing whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

### Extensive Approach

```python
import re
from masri.transcribe.num2text import num2text

chars_to_remove_regex = r'[\*\\,\?\.\!\;\:\"\€\{\}\#\|\^\=\“\%\‘\”\\»\«\/\(\)\[\]]'

def clean_for_kenlm(text):
    text = text.lower()

    #Removing urls
    url_regex = r'https?://\S+|www\.\S+'
    text = re.sub(url_regex, ' ', text)

    # 1. REMOVE BRACKETS AND THEIR CONTENTS COMPLETELY
    # This turns "Id-disponibbiltà (skont il-liġi)" -> "Id-disponibbiltà"
    text = re.sub(r'[\[\]\(\)]', '', text)

    # Removing /a or /i
    text = re.sub(r'/[ai]\b', '', text)

    text = re.sub(r'\b\d+/\d+(/[a-z]+)?\b', ' ', text)

    # This handles any leftover slashes between words (u/jew -> u jew)
    text = text.replace('/', ' ')

    # 2. REMOVE SECTION HEADERS (2.3.6 etc)
    text = re.sub(r'\d+\.\d+(\.\d+)+', '', text)

    # 3. SMART MALTESE SPACING (Fixing "l - iskola" but not "kelma - kelma")
    # This prevents joining long words into "frankenwords"
    # Prefix: Any 1-4 letter Maltese word OR specifically "għall"
    # Suffix: Letters or numbers
    text = re.sub(r'\b([a-zċġħżàùáèéìòć]{1,4}|għall)\s*-\s*([a-zċġħżàùáèéìòć0-9]+)', r'\1-\2', text)

    # Now, treat any REMAINING dashes (those between long words) as spaces
    text = re.sub(r'\s+-\s+', ' ', text)

    # 4. CONVERT STANDALONE NUMBERS
    def replace_num(match):
        prefix_char = match.group(1) # Captures the char before the number (like '-')
        val = match.group(2).strip()
        suffix = match.group(3) if match.group(3) else ""
        
        is_currency = '€' in val
        clean_val = val.replace('€', '').strip()

        # Mapping for 2-10 short forms
        short_forms = {
            2: "żewġ", 3: "tlett", 4: "erba'", 5: "ħames", 
            6: "sitt", 7: "seba'", 8: "tminn", 9: "disa'", 10: "għaxar"
        }

        def format_maltese_unit(n_str, singular, plural):
            n = int(n_str)
            if n == 0: return ""
            if n == 1: return singular
            
            # 11-19 Rule
            if 11 <= n <= 19:
                return f"{num2text(n)}-il {singular}"
            
            # 2-10 Rule
            if n in short_forms:
                return f"{short_forms[n]} {plural}"
            
            # Hundreds Rule
            if n == 100:
                return f"mitt {plural}"
            if n == 200:
                return f"mitejn {plural}"
            if 300 <= n <= 900 and n % 100 == 0:
                prefix_digit = n // 100
                return f"{short_forms.get(prefix_digit, num2text(prefix_digit))} mitt {plural}"
            
            # 20+ or other
            return f"{num2text(n)} {singular}"

        try:
            result = ""
            if suffix == "-il":
                result = f"{num2text(int(clean_val))}-il"
            elif '.' in clean_val:
                parts = clean_val.split('.')
                whole, cents = parts[0], parts[1].ljust(2, '0')[:2]
                if is_currency:
                    e_part = format_maltese_unit(whole, "ewro", "ewro")
                    c_part = format_maltese_unit(cents, "ċenteżmu", "ċenteżmi")
                    result = f"{e_part} u {c_part}" if e_part and c_part else (e_part or c_part)
                else:
                    result = f"{num2text(int(whole))} punt {num2text(int(cents))}"
            elif is_currency:
                result = format_maltese_unit(clean_val, 'ewro', 'ewro')
            else:
                result = num2text(int(clean_val))

            clean_result = result.strip()

            if prefix_char == "-":
                # Attached to a hyphen (e.g., fl-), keep it attached
                return f"-{clean_result} "
            elif prefix_char.isspace():
                # It was a space, keep it as a space
                return f" {clean_result} "
            else:
                # No prefix captured (start of string), just return the word
                return f"{clean_result} "
                
        except:
            return " "


    # UPDATED REGEX: Added ([-\s]?) at the start to catch hyphens or spaces
    text = re.sub(r'([-\s]?)(€?\s?\d+(?:\.\d+)?)(-il)?', replace_num, text)
    
    # 5. FINAL CHARACTER STRIPPING & WHITESPACE
    dashes_to_remove = r'[\u2013\u2014]'
    text = re.sub(dashes_to_remove, ' ', text)
    
    # text = re.sub(chars_to_remove_regex, '', text)
    text = re.sub(r'[^a-z0-9ċġħżàùáèéìòć’`\-\']', ' ', text)

    # 6. SMART QUOTE REMOVAL
    # Rule A: Remove opening quotes, BUT NOT if followed by 'l' (prevents 'l -> l)
    # This preserves: 'l-isfel, 'l hemm
    text = re.sub(r"(^|\s)'+(?!(l\b|il\b))", r"\1", text)

    # Rule B: Remove quotes that have a space after them, 
    # BUT ONLY if they aren't preceded by a Maltese letter.
    # This preserves: jista', ta', b'
    # This removes: sema' , ' 
    text = re.sub(r"(?<![a-zċġħżà])'+(\s|$)", r"\1", text)

    #Fixing x' inhu etc
    text = re.sub(r"\b([fbxtms]')\s+", r"\1", text)
    
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

### Combined Approach

```python
def combined_approach(text, tokenizer_obj=None):
    # --- STEP 1: Rigorous Cleaning ---
    # This handles num2text, Maltese dashes, and smart symbol removal
    cleaned_text = clean_for_kenlm(text)
    
    # --- STEP 2: Tokenization ---
    # If a tokeniser is provided, use it. 
    if tokenizer_obj:
        tokens = tokenizer_obj.tokenize(cleaned_text)
        
        # --- STEP 3: Final Validation ---
        # Removes empty strings or whitespace-only tokens created by the split
        processed_tokens = [t for t in tokens if t.strip()]
        
        # Return as a space-separated string (Ideal for KenLM)
        return " ".join(processed_tokens)
    
    # If no tokeniser is provided, just return the rigorously cleaned string
    return cleaned_text
```

### Processing the Corpus

```python
import multiprocessing
import os

def process_single_entry(entry):
    """
    Worker function: This runs on its own CPU core.
    It takes one entry and returns a list of cleaned lines.
    """
    results = []
    # Handle if entry is a string or a list of sentences
    lines = entry["text"] if isinstance(entry["text"], list) else [entry["text"]]
    for line in lines:
        if line:
            cleaned = regex_approach(line)
            if cleaned:
                results.append(cleaned)
    return results

if __name__ == "__main__":
    # Create the pool
    with multiprocessing.Pool(processes=os.cpu_count()) as pool:
        # PUSHING CHUNKSIZE: This is the lever for CPU usage
        # Increasing this reduces the 'bottleneck' of sending data between cores
        all_results = pool.imap_unordered(process_single_entry, dataset, chunksize=2500)
        
        with open("corpus_untokenised.txt", "w", encoding="utf-8") as f:
            for i, result_list in enumerate(all_results):
                if result_list:
                    f.write("\n".join(result_list) + "\n")
                
                if i % 100000 == 0:
                    print(f"Processed {i} entries...")
```

```python
with open("tok_pros.arpa", "r") as read_file, open("2_untok_red.arpa", "w") as write_file:
  has_added_eos = False
  for line in read_file:
    if not has_added_eos and "ngram 1=" in line:
      count=line.strip().split("=")[-1]
      write_file.write(line.replace(f"{count}", f"{int(count)+1}"))
    elif not has_added_eos and "<s>" in line:
      write_file.write(line)
      write_file.write(line.replace("<s>", "</s>"))
      has_added_eos = True
    else:
      write_file.write(line)
```
